```python
# ============================================================
# 1. IMPORT LIBRARY
# ============================================================
import pandas as pd
import re
import string
import numpy as np
import torch
from tqdm import tqdm
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
from transformers import AutoTokenizer, AutoModel
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from transformers import Trainer, TrainingArguments
import nltk

nltk.download('punkt')
nltk.download('stopwords')

# ============================================================
# 2. CLEANING TEKS
# ============================================================
factory = StemmerFactory()
stemmer = factory.create_stemmer()
stop_words = set(stopwords.words('indonesian'))

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    tokens = word_tokenize(text)
    filtered = [w for w in tokens if w not in stop_words]
    stemmed = [stemmer.stem(w) for w in filtered]
    return ' '.join(stemmed)

# ============================================================
# 3. LOAD DATASET PARFUM
# ============================================================
df = pd.read_excel("Dataset Text Web.xlsx", sheet_name="REKAP")
df = df[['review', 'label']].dropna()
df['clean_review'] = df['review'].astype(str).apply(clean_text)
df['label'] = df['label'].astype(int)
print("Jumlah data:", len(df))
print(df.head())

# ============================================================
# 4. LOAD TOKENIZER DAN MODEL INDOBERT
# ============================================================
model_name = "indobenchmark/indobert-base-p1"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

# ============================================================
# 5. CONVERT TEKS KE EMBEDDING INDO-BERT
# ============================================================
def get_bert_embedding(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=128)
    with torch.no_grad():
        outputs = model(**inputs)
    cls_vec = outputs.last_hidden_state[:, 0, :]
    mean_vec = outputs.last_hidden_state.mean(dim=1)
    emb = (cls_vec + mean_vec) / 2
    return emb.numpy().flatten()


print(" Mengubah review jadi embedding IndoBERT... (butuh beberapa menit)")
embeddings = np.array([get_bert_embedding(t) for t in tqdm(df['clean_review'])])

# ============================================================
# 6. SPLIT TRAIN/TEST
# ============================================================
X_train, X_test, y_train, y_test = train_test_split(
    embeddings, df['label'], test_size=0.2, stratify=df['label'], random_state=42
)

# ============================================================
# 7. OVERSAMPLING UNTUK BALANCING DATA
# ============================================================
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=42)
X_train_res, y_train_res = ros.fit_resample(X_train, y_train)

# ============================================================
# 8. GRID SEARCH UNTUK MENCARI PARAMETER TERBAIK SVM
# ============================================================
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

params = {
    'C': [0.1, 1, 5],
    'kernel': ['linear', 'rbf'],
    'class_weight': [None, 'balanced']
}

print(" Melakukan grid search untuk mencari parameter terbaik...")
grid = GridSearchCV(SVC(), params, cv=3, scoring='accuracy', n_jobs=-1)
grid.fit(X_train_res, y_train_res)

print("Best params:", grid.best_params_)
best_svm = grid.best_estimator_

# ============================================================
# 9. EVALUASI MODEL HASIL GRID SEARCH
# ============================================================
from sklearn.metrics import classification_report, confusion_matrix

y_pred = best_svm.predict(X_test)

print("\n=== CLASSIFICATION REPORT ===")
print(classification_report(y_test, y_pred, target_names=['Negatif', 'Netral', 'Positif']))

print("\n=== CONFUSION MATRIX ===")
print(confusion_matrix(y_test, y_pred))

# ============================================================
# 9. PREDIKSI REVIEW BARU
# ============================================================
def predict_sentiment(text):
    cleaned = clean_text(text)
    emb = get_bert_embedding(cleaned).reshape(1, -1)
    pred = best_svm.predict(emb)[0]
    label = ["Negatif ", "Netral ", "Positif "][pred]
    return label

print("\n=== Prediksi Review Baru ===")
new_reviews = [
    "Bau banget kayak minyak gosok",
    "Wanginya lembut banget dan tahan lama",
    "Biasa aja sih, botolnya kecil tapi lucu",
    "Awalnya ga enak tapi makin lama makin wangi"
]

for text in new_reviews:
    print(f"{text}  {predict_sentiment(text)}")
```

```python
# ============================================================
# 1. IMPORT LIBRARY
# ============================================================
import pandas as pd
import re
import numpy as np
import torch
from tqdm import tqdm
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.svm import SVC
from imblearn.over_sampling import RandomOverSampler
from transformers import AutoTokenizer, AutoModel
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter
import nltk

nltk.download('punkt')
nltk.download('stopwords')

# ============================================================
# 2. CLEANING TEKS
# ============================================================
factory = StemmerFactory()
stemmer = factory.create_stemmer()
stop_words = set(stopwords.words('indonesian'))

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    tokens = word_tokenize(text)
    filtered = [w for w in tokens if w not in stop_words]
    stemmed = [stemmer.stem(w) for w in filtered]
    return ' '.join(stemmed)

# ============================================================
# 3. LOAD DATASET PARFUM
# ============================================================
df = pd.read_excel("Dataset Text Web.xlsx", sheet_name="REKAP")
df = df[['review', 'label']].dropna()
df['clean_review'] = df['review'].apply(clean_text)
df['label'] = df['label'].astype(int)
print("Jumlah data:", len(df))
print(df.head())

# ============================================================
# 4. LOAD TOKENIZER DAN MODEL INDOBERT
# ============================================================
model_name = "indobenchmark/indobert-base-p1"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

# ============================================================
# 5. CONVERT TEKS KE EMBEDDING DENGAN MEAN POOLING
# ============================================================
def get_bert_embedding(text):
    inputs = tokenizer(
        text,
        return_tensors="pt",
        truncation=True,
        padding=True,
        max_length=128
    )
    with torch.no_grad():
        outputs = model(**inputs)
        last_hidden = outputs.last_hidden_state

    # CLS token
    cls_vec = last_hidden[:, 0, :]

    # Mean pooling dengan masking (biar padding tidak ikut dihitung)
    attention_mask = inputs["attention_mask"]
    mask_expanded = attention_mask.unsqueeze(-1).expand(last_hidden.size()).float()
    sum_embeddings = torch.sum(last_hidden * mask_expanded, dim=1)
    sum_mask = torch.clamp(mask_expanded.sum(dim=1), min=1e-9)
    mean_vec = sum_embeddings / sum_mask

    # Gabungan hybrid
    emb = (cls_vec + mean_vec) / 2

    # Normalisasi vektor
    emb = emb / emb.norm(dim=1, keepdim=True)

    return emb.numpy().flatten()


print("\n Mengubah review jadi embedding IndoBERT (mean pooling)...")
embeddings = np.array([get_bert_embedding(t) for t in tqdm(df['clean_review'])])

# ============================================================
# 6. SPLIT TRAIN/TEST
# ============================================================
X_train, X_test, y_train, y_test = train_test_split(
    embeddings, df['label'], test_size=0.2, stratify=df['label'], random_state=42
)

# ============================================================
# 7. OVERSAMPLING UNTUK BALANCING DATA
# ============================================================
ros = RandomOverSampler(random_state=42)
X_train_res, y_train_res = ros.fit_resample(X_train, y_train)
print("\n=== Distribusi Data Setelah Oversampling ===")
print(Counter(y_train_res))


# ============================================================
# 8. GRID SEARCH UNTUK PARAMETER TERBAIK SVM (OPTIMIZED)
# ============================================================
# params = {
#     'C': [0.01, 0.1, 1, 5, 10],
#     'kernel': ['linear', 'rbf', 'poly'],
#     'gamma': ['scale', 'auto'],
#     'class_weight': [None, 'balanced']
# }
params = {
    'C': [8, 10, 12],
    'gamma': [0.3, 0.4, 0.5],
    'class_weight': ['balanced']
}
print("\n Melakukan Grid Search SVM (5-fold cross-validation)...")
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
grid = GridSearchCV(SVC(kernel='rbf', probability=True), params, cv=5, scoring='f1_macro', n_jobs=-1)
grid.fit(X_train_res, y_train_res)
print("Best params:", grid.best_params_)

svm_model = grid.best_estimator_
svm_model = SVC(
    kernel='rbf',
    C=12,               # Lebih besar  keputusan lebih tajam
    gamma=0.5,          # Fokus ke fitur spesifik (kata negatif, netral)
    class_weight='balanced',
    probability=True,
    random_state=42
)
svm_model.fit(X_train_res, y_train_res)
# ============================================================
# 9. EVALUASI MODEL
# ============================================================
y_pred = svm_model.predict(X_test)
print("\n=== CLASSIFICATION REPORT ===")
print(classification_report(y_test, y_pred, target_names=['Negatif', 'Netral', 'Positif']))

print("\n=== CONFUSION MATRIX ===")
print(confusion_matrix(y_test, y_pred))

# ============================================================
# 10. UJI COBA REVIEW MANUAL / RANDOM
# ============================================================
label_map = {0: "Negatif ", 1: "Netral ", 2: "Positif "}

def predict_review(text):
    clean = clean_text(text)
    emb = get_bert_embedding(clean).reshape(1, -1)
    probs = svm_model.predict_proba(emb)
    pred = np.argmax(probs)
    conf = np.max(probs)
    print("", text)
    print(f" Output : {label_map[pred]} (Confidence: {conf:.2f})")
    print("="*60)

print("\n===  UJI 10 DATA RANDOM REVIEW PARFUM ===\n")

sample_texts = [
    "Wanginya enak banget, elegan dan tahan lama.",
    "Aromanya segar tapi cepat hilang.",
    "Bikin pusing, baunya terlalu kuat.",
    "Botolnya kecil tapi desainnya lucu.",
    "Harumnya bikin percaya diri, recommended!",
    "Kurang suka, aromanya aneh di kulit.",
    "Standar aja sih, nggak terlalu spesial.",
    "Mantap banget, wanginya eksklusif dan mewah.",
    "Oke lah untuk harga segitu.",
    "Bau seperti obat, nggak enak sama sekali."
]

for t in sample_texts:
    predict_review(t)

```
