```python
# ============================================================
# 1. IMPORT LIBRARY
# ============================================================
import pandas as pd
import re
import string
import numpy as np
import torch
from tqdm import tqdm
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
from transformers import AutoTokenizer, AutoModel
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from transformers import Trainer, TrainingArguments
import nltk

nltk.download('punkt')
nltk.download('stopwords')

# ============================================================
# 2. CLEANING TEKS
# ============================================================
factory = StemmerFactory()
stemmer = factory.create_stemmer()
stop_words = set(stopwords.words('indonesian'))

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    tokens = word_tokenize(text)
    filtered = [w for w in tokens if w not in stop_words]
    stemmed = [stemmer.stem(w) for w in filtered]
    return ' '.join(stemmed)

# ============================================================
# 3. LOAD DATASET PARFUM
# ============================================================
df = pd.read_excel("Dataset Text Web.xlsx", sheet_name="REKAP")
df = df[['review', 'label']].dropna()
df['clean_review'] = df['review'].astype(str).apply(clean_text)
df['label'] = df['label'].astype(int)
print("Jumlah data:", len(df))
print(df.head())

# ============================================================
# 4. LOAD TOKENIZER DAN MODEL INDOBERT
# ============================================================
model_name = "indobenchmark/indobert-base-p1"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

# ============================================================
# 5. CONVERT TEKS KE EMBEDDING INDO-BERT
# ============================================================
def get_bert_embedding(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=128)
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state[:, 0, :].numpy().flatten()

print("🔄 Mengubah review jadi embedding IndoBERT... (butuh beberapa menit)")
embeddings = np.array([get_bert_embedding(t) for t in tqdm(df['clean_review'])])

# ============================================================
# 6. SPLIT TRAIN/TEST
# ============================================================
X_train, X_test, y_train, y_test = train_test_split(
    embeddings, df['label'], test_size=0.2, stratify=df['label'], random_state=42
)

# ============================================================
# 7. OVERSAMPLING UNTUK BALANCING DATA
# ============================================================
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=42)
X_train_res, y_train_res = ros.fit_resample(X_train, y_train)

# ============================================================
# 8. GRID SEARCH UNTUK MENCARI PARAMETER TERBAIK SVM
# ============================================================
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

params = {
    'C': [0.1, 1, 5],
    'kernel': ['linear', 'rbf'],
    'class_weight': [None, 'balanced']
}

print("🔍 Melakukan grid search untuk mencari parameter terbaik...")
grid = GridSearchCV(SVC(), params, cv=3, scoring='accuracy', n_jobs=-1)
grid.fit(X_train_res, y_train_res)

print("Best params:", grid.best_params_)
best_svm = grid.best_estimator_

# ============================================================
# 9. EVALUASI MODEL HASIL GRID SEARCH
# ============================================================
from sklearn.metrics import classification_report, confusion_matrix

y_pred = best_svm.predict(X_test)

print("\n=== CLASSIFICATION REPORT ===")
print(classification_report(y_test, y_pred, target_names=['Negatif', 'Netral', 'Positif']))

print("\n=== CONFUSION MATRIX ===")
print(confusion_matrix(y_test, y_pred))

# ============================================================
# 9. PREDIKSI REVIEW BARU
# ============================================================
def predict_sentiment(text):
    cleaned = clean_text(text)
    emb = get_bert_embedding(cleaned).reshape(1, -1)
    pred = svm_model.predict(emb)[0]
    label = ["Negatif 😡", "Netral 😐", "Positif 😊"][pred]
    return label

print("\n=== Prediksi Review Baru ===")
new_reviews = [
    "Bau banget kayak minyak gosok",
    "Wanginya lembut banget dan tahan lama",
    "Biasa aja sih, botolnya kecil tapi lucu",
    "Awalnya ga enak tapi makin lama makin wangi"
]

for text in new_reviews:
    print(f"{text} → {predict_sentiment(text)}")
```

```python
# ============================================================
# 1. IMPORT LIBRARY
# ============================================================
import pandas as pd
import re
import numpy as np
import torch
from tqdm import tqdm
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.svm import SVC
from imblearn.over_sampling import RandomOverSampler
from transformers import AutoTokenizer, AutoModel
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk

nltk.download('punkt')
nltk.download('stopwords')

# ============================================================
# 2. CLEANING TEKS
# ============================================================
factory = StemmerFactory()
stemmer = factory.create_stemmer()
stop_words = set(stopwords.words('indonesian'))

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    tokens = word_tokenize(text)
    filtered = [w for w in tokens if w not in stop_words]
    stemmed = [stemmer.stem(w) for w in filtered]
    return ' '.join(stemmed)

# ============================================================
# 3. LOAD DATASET PARFUM
# ============================================================
df = pd.read_excel("Dataset Text Web.xlsx", sheet_name="REKAP")
df = df[['review', 'label']].dropna()
df['clean_review'] = df['review'].apply(clean_text)
df['label'] = df['label'].astype(int)
print("Jumlah data:", len(df))
print(df.head())

# ============================================================
# 4. LOAD TOKENIZER DAN MODEL INDOBERT
# ============================================================
model_name = "indobenchmark/indobert-base-p1"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

# ============================================================
# 5. CONVERT TEKS KE EMBEDDING DENGAN MEAN POOLING
# ============================================================
def get_bert_embedding(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=128)
    with torch.no_grad():
        outputs = model(**inputs)
    # Mean pooling: rata-rata semua token untuk representasi global
    embedding = outputs.last_hidden_state.mean(dim=1).numpy().flatten()
    return embedding

print("\n🔄 Mengubah review jadi embedding IndoBERT (mean pooling)...")
embeddings = np.array([get_bert_embedding(t) for t in tqdm(df['clean_review'])])

# ============================================================
# 6. SPLIT TRAIN/TEST
# ============================================================
X_train, X_test, y_train, y_test = train_test_split(
    embeddings, df['label'], test_size=0.2, stratify=df['label'], random_state=42
)

# ============================================================
# 7. OVERSAMPLING UNTUK BALANCING DATA
# ============================================================
ros = RandomOverSampler(random_state=42)
X_train_res, y_train_res = ros.fit_resample(X_train, y_train)

# ============================================================
# 8. GRID SEARCH UNTUK PARAMETER TERBAIK SVM (OPTIMIZED)
# ============================================================
params = {
    'C': [0.01, 0.1, 1, 5, 10],
    'kernel': ['linear', 'rbf', 'poly'],
    'gamma': ['scale', 'auto'],
    'class_weight': [None, 'balanced']
}

print("\n🔍 Melakukan Grid Search SVM (5-fold cross-validation)...")
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
grid = GridSearchCV(SVC(), params, cv=cv, scoring='accuracy', n_jobs=-1)
grid.fit(X_train_res, y_train_res)

print("\n✅ Best params:", grid.best_params_)
svm_model = grid.best_estimator_

# ============================================================
# 9. EVALUASI MODEL
# ============================================================
y_pred = svm_model.predict(X_test)
print("\n=== CLASSIFICATION REPORT ===")
print(classification_report(y_test, y_pred, target_names=['Negatif', 'Netral', 'Positif']))
print("\n=== CONFUSION MATRIX ===")
print(confusion_matrix(y_test, y_pred))

```

```python
# ============================================================
# 10. VISUALISASI HASIL EVALUASI
# ============================================================
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score

# Hitung akurasi total
acc = accuracy_score(y_test, y_pred)
print(f"\n🎯 Akurasi Total: {acc * 100:.2f}%")

# Buat confusion matrix dalam bentuk dataframe
cm = confusion_matrix(y_test, y_pred)
labels = ["Negatif (0)", "Netral (1)", "Positif (2)"]

# Visualisasi confusion matrix
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
plt.title(f"Confusion Matrix (Akurasi: {acc*100:.2f}%)")
plt.xlabel("Prediksi")
plt.ylabel("Aktual")
plt.show()

# ============================================================
# 11. Grafik Perbandingan Prediksi vs Aktual
# ============================================================
result_df = pd.DataFrame({"Actual": y_test, "Predicted": y_pred})
plt.figure(figsize=(8,4))
sns.countplot(data=result_df, x="Actual", palette="cool", label="Aktual")
sns.countplot(data=result_df, x="Predicted", palette="Reds", label="Prediksi", alpha=0.6)
plt.legend()
plt.title("Distribusi Label Aktual vs Prediksi")
plt.show()
```
